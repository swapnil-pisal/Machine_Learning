```python
# Step 1: Lowercasting + Remove punctuation 

import re
text = "I loved the new phone!!! It's absolutely amazing."

# convert to lowercase 
text = text.lower()

# Remove punctuation 
text = re.sub(r'[^\w\s]', '',text)
print(text)
```
